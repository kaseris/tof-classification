```python
import os
import numpy as np

raw_data_path = '/home/kaseris/captures'
output_folder = 'preprocessed_data'

classes = os.listdir(raw_data_path)

gesture_to_idx = {'comehere': 0, 'spin': 1, 'stop': 2}

for c in classes:
    path = os.path.join(raw_data_path, c)
    samples = os.listdir(os.path.join(raw_data_path, c))
    
    for sample_id in samples:
        full_path = os.path.join(path, sample_id)
        
        sensors = ['0.csv', '1.csv', '2.csv']
        sensor_data = []
        for fname in sensors:
            
            file_path = os.path.join(full_path, fname)
            print(f'Processing file: {file_path}')
            data = np.genfromtxt(file_path, delimiter=',').astype(np.float32)
            data = data / 255.0
            sensor_data.append(data)
        sensor_data = np.stack(sensor_data, axis=1)
        label = gesture_to_idx[c]
        out_file = os.path.join(output_folder, f'{c}_{sample_id}.npz')
        np.savez(out_file, array=sensor_data, label=label)
```
